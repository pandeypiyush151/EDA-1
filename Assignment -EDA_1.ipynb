#### Checking for Null Values

```python
df.isnull().sum()
```

```python


# Impute missing values with the median
df = df.fillna(df.median())

# Alternatively, drop rows with missing values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python


# Identify columns with mixed data types
mixed_cols = df.select_dtypes(include=['object']).columns[df.select_dtypes(include=['object']).applymap(lambda x: x.isdigit()).any()]

# Convert mixed columns to numeric
for col in mixed_cols:
  df[col] = pd.to_numeric(df[col], errors='coerce')

# Verify data types after conversion
df.dtypes
```

```python


# Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Loop through numerical columns and treat outliers using IQR method
for col in num_cols:
  # Calculate IQR
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1

  # Define lower and upper bounds
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Replace outliers with bounds
  df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Verify outlier treatment
df[num_cols].describe()
```

```python


# Create histograms for numerical variables
df.hist(figsize=(10, 8), bins=10)
plt.show()

# Create boxplots for numerical variables
df.boxplot(figsize=(10, 8))
plt.show()
```

```python




# Generate a correlation heatmap
corr_matrix = df[['LB', 'AC', 'FM', 'UC', 'DL']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python

# Create a pairplot with density plots on the diagonal
sns.pairplot(df[['LB', 'AC', 'FM', 'UC', 'DL']], diag_kind='kde')
plt.show()

# Create a violin plot for each numerical variable
for col in num_cols:
  sns.violinplot(y=col, data=df)
  plt.show()
```
